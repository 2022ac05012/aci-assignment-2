```python
import random

# Define the crossword grid
grid = [
    ['#', 'D', 'U', 'C', 'K', '#', 'D'],
    ['E', '#', '#', 'R', '#', '#', 'O'],
    ['M', '#', '#', 'O', '#', '#', 'V'],
    ['U', '#', 'S', 'W', 'A', 'N', 'E'],
    ['#', '#', 'P', '#', '#', '#', '#'],
    ['P', 'E', 'A', 'C', 'O', 'C', 'K'],
    ['#', '#', 'R', '#', '#', '#', '#'],
    ['P', 'A', 'R', 'R', 'O', 'T', '#'],
    ['#', '#', 'O', '#', '#', '#', '#'],
    ['#', '#', 'W', 'R', 'E', 'N', '#']
]

# Define the list of words and their corresponding locations
words_and_locations = {
    "DUCK": [(0, 1), (0, 2), (0, 3), (0, 4)],
    "EMU": [(1, 0), (2, 0), (2, 0)],
    "CROW": [(0, 3), (1, 3), (2, 3),(3, 3)],
    "DOVE": [(0, 6), (1, 6), (2, 6),(3,6)],
    "SPARROW": [(3, 2), (4, 2), (5, 2),(6, 2), (7, 2), (8, 2),(9, 2)],
    "SWAN": [(3, 2), (3, 3), (3, 4), (3, 5)],
    "PEACOCK": [(5, 0), (5, 1), (5, 2), (5, 3), (5, 4), (5, 5),(5,6)],
    "PARROT": [(7, 0), (7, 1), (7, 2), (7, 3), (7, 4)],
    "WREN": [(9, 2), (9, 3), (9, 4), (9, 5)]
}

# Initialize player scores
player_score = 0
computer_score = 0

# Define a function to display the grid
def display_grid():
    for row in grid:
        print(" ".join(row))
    print()

# Define a function to check if a word can be placed in the grid at a given location
def can_place_word(word, locations):
    for (row, col) in locations:
        if grid[row][col] != "#" and grid[row][col] != word[locations.index((row, col))]:
            return False
    return True

# Define a function to place a word in the grid
def place_word(word, locations):
    for (row, col) in locations:
        grid[row][col] = word[locations.index((row, col))]

# Define the Min-Max algorithm for the computer player (simplified version)
def minimax(depth, maximizing_player):
    if depth == 0:
        return random.choice(list(words_and_locations.keys())), 0  # Random choice at leaf nodes
    
    if maximizing_player:
        best_word = None
        best_score = -float("inf")
        for word, locations in words_and_locations.items():
            if can_place_word(word, locations):
                place_word(word, locations)
                score = minimax(depth - 1, False)[1]
                place_word("#" * len(word), locations)  # Undo the placement
                if score > best_score:
                    best_word = word
                    best_score = score
        return best_word, best_score
    else:
        best_word = None
        best_score = float("inf")
        for word, locations in words_and_locations.items():
            if can_place_word(word, locations):
                place_word(word, locations)
                score = minimax(depth - 1, True)[1]
                place_word("#" * len(word), locations)  # Undo the placement
                if score < best_score:
                    best_word = word
                    best_score = score
        return best_word, best_score

# Define the main game loop
count =0;
while True:
    display_grid()
    
    # Player's turn
    player_word = input("Player, choose a word to place: ").strip().upper()
    
    if player_word not in words_and_locations or not can_place_word(player_word, words_and_locations[player_word]):
        print("Invalid word. Try again.")
        continue
    
    place_word(player_word, words_and_locations[player_word])
    player_score += len(player_word)
    del words_and_locations[player_word]
    
    # Check for game over
    if not words_and_locations:
        break
    
    display_grid()
    
    # Computer's turn using Min-Max
    computer_word, _ = minimax(2, True)  # Adjust the depth as needed
    computer_locations = words_and_locations[computer_word]
    
    print(f"Computer chooses: {computer_word}")
    place_word(computer_word, computer_locations)
    computer_score += len(computer_word)
    del words_and_locations[computer_word]
    count+=1;
    if (count == 10):
        break;

# Print final scores and grid
print("Game Over!")
display_grid()
print(f"Player Score: {player_score}")
print(f"Computer Score: {computer_score}")
```
